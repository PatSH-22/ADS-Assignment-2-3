# 1. Breakdown of distance from home by job role and attrition.


```python
# Let now groupby by attrition and jobrole then classify based on mean distance from home
df = employee_data.groupby(['Attrition', 'JobRole'])['DistanceFromHome'].mean().reset_index()
df
```

```python
# Graphically showing the above

plt.figure(figsize=(15,10))
sns.barplot(x = 'Attrition', y = 'DistanceFromHome', data = df, hue = 'JobRole' )
plt.title('Attrition Analysis based on Home Distance and Job Role')
plt.xlabel('Attrition')
plt.ylabel('DistanceFromHome')
plt.show()
```

Distance from home somewhat has a ripple effect on attrition. As seen healthcare representative living far from work have a higher attrition.

# 2. Compare Average Monthly Income by Education and Attrition

```python
# To get a glimpse of the minimum, maximumand average monthly income
employee_data['MonthlyIncome'].describe()
```

```python
df_1 = employee_data.groupby(['Attrition', 'Education'])['MonthlyIncome'].mean().reset_index()
df_1
```

```python
# Plotting
sns.barplot(x = 'Education', y = 'MonthlyIncome', data = employee_data, hue = 'Attrition' )
```

As the level of education increases and the monthly income is still low there's a higher chance of attrition. 

```python
# You can also plot
plt.figure(figsize=(15,10))
sns.barplot(x = 'Attrition', y = 'MonthlyIncome', data = df_1, hue = 'Education' )
plt.title('Attrition Analysis based on level of education and monthly income')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.show()
```
